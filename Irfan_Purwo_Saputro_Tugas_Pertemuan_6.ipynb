```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mus = pd.read_csv('mushrooms.csv')
mus.head()
```

```python
y = mus['class']
y = y.map({'p':1, 'e':0})
X1 = mus['cap-shape'].map({'x':0, 'b':1, 's':2, 'f':3,'k':4,'c':5})
X2 = mus['bruises'].map({'t':0, 'f':1})

v1=[]
v2=[]
for x1 in X1:
  v1.append(x1)
for x2 in X2:
  v2.append(x2)

X = pd.DataFrame()

X['cap-shape'] = v1
X['bruises'] = v2
print(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
from sklearn.tree import DecisionTreeClassifier

# Instantiate DecisionTreeClassifier 'dt' dengan kedalaman maksimum 6
dt = DecisionTreeClassifier(max_depth=6, random_state=1)

# Sesuaikan dt ke set training
dt.fit(X_train, y_train)

# Pmemprediksi label set test
y_pred = dt.predict(X_test)
print(y_pred[0:5])
```

```python
from sklearn.metrics import accuracy_score

# Memprediksi test set labels
y_pred = dt.predict(X_test)

#  menghitung set accuracy
acc = accuracy_score(y_test, y_pred)
print("Test set accuracy: {:.2f}".format(acc))
```
